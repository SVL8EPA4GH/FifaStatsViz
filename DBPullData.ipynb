```python
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
from config import engine
from flask import Flask
```

```python
sqlengine = create_engine(engine)
conn = sqlengine.connect()
```

```python
clubs = pd.read_sql('select * from clubs', conn)
clubs.head()
```

```python
player_info = pd.read_sql('select * from player_info', conn)
player_info.head()
```

```python
player_stats = pd.read_sql('select * from player_stats', conn)
player_stats.head()
```

```python
countries = player_info['nationality'].unique()
len(countries)
```

```python
country_medians = player_info.groupby("nationality").median()
country_medians

country_count = player_info.groupby("nationality").count()
country_count
```

```python
country_data = pd.DataFrame({"Median Wages":country_medians['wage'], "Player Count":country_count['fifa_id']})
country_data
```

```python
clubmerge = pd.merge(player_info, clubs, on="club_id")
clubmerge
```

```python
club_wages = clubmerge.groupby("club").median()
club_wages
```

```python
club_count = clubmerge.groupby('club').count()
club_count
```

```python
club_data = pd.DataFrame({"Median Age": club_wages['age'], "Median Wages": club_wages['wage'], "Player Counts": club_count['fifa_id']})
club_data
```

```python
player_stats.columns
```

```python
radar_stats = player_stats[{'fifa_id','acceleration', 'gkreflexes', 'positioning', 'aggression', 'shotpower'}]
radar_stats
```

```python
radarmerge = pd.merge(radar_stats, player_info, on="fifa_id")
radarmerge = radarmerge.drop(axis=1, columns = ['age','nationality','club_id', 'value', 'wage', 'position', 'overall'])

# radarmerge = radarmerge.set_index('ranking')
radarmerge.head(10)
```
